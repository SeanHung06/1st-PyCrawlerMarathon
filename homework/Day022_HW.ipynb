### ① 取出 台北市士林區 2018/01 – 2018/08 的 SO2 資料

```python
# 打開瀏覽器

from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='./chromedriver')

browser.get("http://taqm.epa.gov.tw/taqm/tw/MonthlyAverage.aspx")
```

```python
# 模擬使用者操作行為，選擇/點擊

selectSite = Select(browser.find_element_by_id("ctl05_ddlSite"))
selectSite.select_by_value('11')
selectYear = Select(browser.find_element_by_id("ctl05_ddlYear"))
selectYear.select_by_value('2018')

browser.find_element_by_id('ctl05_btnQuery').click()
```

```python
from datetime import datetime, timedelta

# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source

soup = BeautifulSoup(html_source, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
d = {}
Date_Aug = datetime.strptime('201809', "%Y%m")
Date_Aug_2 = datetime.strftime(Date_Aug,"%Y%m")

for tr in table.find_all("tr")[1:]:
    for i , td in enumerate(tr.find_all("td")) :
        #print(i,td)
        if len(tr.find_all("td")) == 5:
            if (i == 0 ):
                items = td.text
                d.setdefault(items,{})
            if (i == 2) :
                date  = td.text
            if (i == 3):
                data = td.text
                d[items][date]  = data
                
        
        if len(tr.find_all("td")) == 3:
            if (i == 0):
                date  = td.text
            if (i == 1):
                data = td.text
                
                if(date != '\xa0'):
                    date_time = datetime.strptime(date, "%Y/%m")
                    date_time2 = datetime.strftime(date_time,"%Y%m")
                    if (date_time2 < Date_Aug_2) :
                        d[items][date]  = data

    #print(date)

import pandas as pd
df = pd.DataFrame(d)
#pd.DataFrame(d)
d
print(df["SO2"],"\n",df["CO"])
```
