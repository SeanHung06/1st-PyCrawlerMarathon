### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list-2020-01-19-0.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text,d.find(class_="tag"))
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
res = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    date = d.find(class_="date").text
    title = d.find_all('a')[-1].text
    tag = d.find(class_="tag").text
    res.setdefault(tag, [])
    res[tag].append({
    'title': title,
    'date': date 
    })
print(res)
```

### ③ 哪一個類別的文章最多

```python
data = []
for i in res:
    print(i, len(res[i]))
    d = {}
    d['tag'] = i
    d['count'] = len(res[i])
    data.append(d)
for d in sorted(data, key=lambda d: d['count'], reverse = True):
    print(d)

```
